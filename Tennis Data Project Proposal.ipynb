```python
import pandas as pd
tennis_data = pd.read_csv("Tennis_Dataset 1.csv")
tennis_data.head()
```

```python
tennis_data_2 = tennis_data.loc[:, ["winner_rank_points", "winner_ht"]]
tennis_data_2.head()
```

```python
tennis_data_3=tennis_data_2.dropna(axis=1)
tennis_data_3.head()

```

```python
import altair as alt

x = 'winner_rank_points'
y = 'winner_ht'

plot = alt.Chart(tennis_data_3).mark_point().encode(
    x = x,
    y = y
).properties(
    title = "Winners' rank points and their corressponding Heights"
)
plot
```
